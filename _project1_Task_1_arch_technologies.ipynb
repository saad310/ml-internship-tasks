```python
from ultralytics import YOLO
from ultralytics import SAM

#Load the SAM model
yolo_model=YOLO("/content/drive/MyDrive/Colab Notebooks/runs/detect/train3/weights/best.pt")

#run the batched inference on a list of images
results =yolo_model("/content/drive/MyDrive/Colab Notebooks/TumorDetection/test_images/pituitary_960_jpg.rf.b21c9456dc94792be07f04c9f6cd8079.jpg")
#load the sam model/take its version
sam_model=SAM("sam2_b.pt")
for result in results:
  class_ids=result.boxes.cls.int().tolist() #noqa
  if len(class_ids):
    boxes = result.boxes.xyxy #boxes object for bbox outputs
    sam_results=sam_model(result.orig_img,bboxes=boxes,verbose=False,save=True,device=0)
```
